```python
#checkig for missing data

Total= df.isnull().sum().sort_values(ascending=False)
Percent = (Total/df.shape[0])*100
Datatypes = df.dtypes

pd.concat([Total,Percent,Datatypes], axis=1, keys=["Total","Percent","Datatypes"]).sort_values(by=["Total"], ascending=False)
```

```python
# Filling the missing values

# defining the columns to fill
cols_to_fill = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5'] 

for col in cols_to_fill:
  df[col] = df[col].fillna(df[col].mean())
```

```python
# There are negative values for sales whic cannot be true. We therefore need to drop any values less than 0

#Removing values less than 0

df.loc[df["Weekly_Sales"] < 0 , "Weekly_Sales" ] = 0
```
